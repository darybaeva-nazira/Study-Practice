```python
"""с запросом имени файла"""
def read_file():
    while True:
        filename = input("Введите имя файла или 'exit' для выхода: ")
        if filename == 'exit':
            print("Выход из программы.")
            break
        try:
            file = open(filename+".txt", "r")
            for line in file:
               print(line)
            file.close()
        except FileNotFoundError:
            print("the file that you would like to read is not found")
        except:
            print("some error occured")
```

```python
"""
Создайте класс Users со следующими атрибутами и функциями:
a) имя
b) фамилия
c) год рождения
d) адрес электронной почты
e) getAge(): возвращает возраст пользователя, принимая во внимание, что текущий год 2023
"""
```

```python
class Users:
    def __init__(self, name, lastname, bithyear, email):
        self.name = name
        self.lastname = lastname
        self.bithyear = bithyear
        self.email = email
        self.grades = []

    def get_age(self):
        return 2023-int(self.bithyear)
```

```python
"""
Преобразовать каждую строку файла в объект класса Users и сохранить их все в списке. 
Чтобы разделить слова, разделенные запятой, используйте функцию split(). 
Например, если: s = "a,b,c", тогда t = s.split(',') вернет ['a', 'b', 'c']. 
Таким образом разделите данные в каждой строке на отдельные части и присвойте эти значения атрибутам класса «Users».
"""
```

```python
try:
    file = open("users.txt", "r")

    users_list = []

    for line in file:
        t = line.split(',')
        user = Users(t[1], t[2], t[4], t[3])
        users_list.append(user)
    file.close()

    for user in users_list:
        print(user.name, user.lastname, user.email)

except FileNotFoundError:
    print("the file that you would like to read is not found")

except:
    print("some error occured")
```

```python
"""
написать функцию filter(low, high, users), которая будет принимать параметры по умолчанию "low"=30, "high"=50 и 
простой параметр - список объектов класса Users. Функция должна пройтись по всем пользователям и проверить возраст с параметрами low и high. 
Мы должны вернуть только тех пользователей, чей возраст выше нижнего параметра low и ниже верхнего параметра high. 
Распечатайте их адрес электронной почты и возраст. Пожалуйста, разделите данные (адрес электронной почты и возраст) запятой. 
Затем вызовите эту функцию.
"""
```

```python
def filter(users, low=30, high=50):

    for user in users:

        age = user.get_age()

        if age > low and age < high:
            print(user.email, age, sep=",")
```
