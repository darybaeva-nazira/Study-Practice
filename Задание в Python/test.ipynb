```python
def read_file():
    while True:
        filename = input("Введите имя файла или 'exit' для выхода: ")
        if filename == 'exit':
            print("Выход из программы.")
            break
        try:
            file = open(filename+".txt", "r")
            for line in file:
               print(line)
            file.close()
        except FileNotFoundError:
            print("the file that you would like to read is not found")
        except:
            print("some error occured")
```

```python
read_file()
```

```python
class Users:
    def __init__(self, id, name, lastname, email, bithyear):
        self.id = id
        self.name = name
        self.lastname = lastname
        self.email = email
        self.bithyear = int(bithyear)

    def get_age(self):
        return 2023 - self.bithyear
```

```python
users_list = []

try:
    file = open("users.txt", "r")

    for line in file:
        line = line.strip()

        t = line.split(',')

        user = Users(t[0], t[1], t[2], t[3], t[4])

        users_list.append(user)

    file.close()

except FileNotFoundError:
    print("the file that you would like to read is not found")

except:
    print("some error occured")
```

```python
def filter(users, low=30, high=50):

    for user in users:

        age = user.get_age()

        if age > low and age < high:
            print(user.email, age, sep=",")
```

```python
filter(users_list)
```
