## Вопрос 1. Набор данных включает как отмененные, так и завершенные рейсы. Определите, есть ли больше выполненных или отмененных рейсов, и посчитайте числовую разницу между ними.

```python
flight_counts = flights_df['Cancelled'].value_counts()

completed = flight_counts[0]
cancelled = flight_counts[1]

difference = abs(completed - cancelled)

print('Выполненных рейсов:', completed)
print('Отменённых рейсов:', cancelled)
print('Разница:', difference)
if completed > cancelled:
    print('Выполненных рейсов больше на', completed - cancelled)
else:
    print('Отменённых рейсов больше на', cancelled - completed)
```

## Вопрос 4. Давайте рассмотрим время отправления и рассмотрим распределение по часам (столбец DepHour, который мы создали ранее). В какой час самый высокий процент рейсов?

```python
hour_procent = flights_df['DepHour'].value_counts(normalize=True) * 100
print(hour_procent.sort_values(ascending=False))
print("Час:", hour_procent.index[0])
print("Процент рейсов:", round(hour_procent.iloc[0], 2), "%")
```

## Вопрос 5. Хорошо, теперь давайте рассмотрим распределение отмененных рейсов по времени. В какой час наименьший процент отмененных рейсов?

```python
# Всего рейсов по каждому часу
all_flights = flights_df.groupby('DepHour').size()

# Отменённых рейсов по каждому часу
cancelled_flights = flights_df.groupby('DepHour')['Cancelled'].sum()

# Процент отмен
cancelled_percent = cancelled_flights / all_flights * 100

print(cancelled_percent.sort_values())
print("Час:", cancelled_percent.idxmin())
print("Процент отмен:", round(cancelled_percent.min(), 2), "%")
```

## Вопрос 7. Найдите самый загруженный час или, другими словами, час, когда количество вылетающих рейсов достигает максимума.

```python
hour_counts = flights_df['DepHour'].value_counts().sort_index()

print("Cамый загруженный час", hour_counts.idxmax())
print("Kоличество вылетающих рейсов", hour_counts.max())
```

## Вопрос 8. Поскольку мы знаем час отправления, было бы интересно изучить среднюю задержку для соответствующего часа. Бывают ли случаи, когда самолеты в среднем вылетали раньше положенного? И если да, то в какое время это произошло?

```python
hour_delay = flights_df.groupby('DepHour')['DepDelay'].mean()
early_hours = hour_delay[hour_delay < 0]

print("Kакое время это произошло", early_hours.idxmin())
print("Cреднюю задержку для соответствующего часа", early_hours.min())
```

## Вопрос 9. Учитывая только выполненные перевозчиком рейсы, которые вы нашли в вопросе 3, найдите распределение этих рейсов по часам. В какое время вылетает наибольшее количество самолетов?

## Вопрос 11. Постройте распределение причин отмены рейса (CancellationCode).

Какая наиболее частая причина отмены рейса?

Перевозчик
Погодные условия
Национальная воздушная система
Причины безопасности

```python
cancellation_list = flights_df['CancellationCode'].value_counts().sort_index()
print(cancellation_list)
cancellation_descriptions = {
    'A': 'Carrier',
    'B': 'Weather',
    'C': 'National Air System',
    'D': 'Security'
}
cancellation_list_named = cancellation_list.rename(index=cancellation_descriptions)
print(cancellation_list_named)

print("Наибольее распрастраненная причина отмены рейса", cancellation_list_named.idxmax())
```

## Вопрос 12. Какой маршрут наиболее частый по количеству рейсов?

(Обратите внимание на столбцы 'Origin' и 'Dest'. Рассматривайте направления A->B и B->A как разные маршруты)

Нью-Йорк – Вашингтон (JFK-IAD)
Сан-Франциско – Лос-Анджелес (SFO-LAX)

Подсказка: учитывайте только положительные задержки
449

```python
delayed = flights_df[flights_df['DepDelay'] > 0]
print(delayed['Route'].value_counts().head(5))
top_routes_list = delayed['Route'].value_counts().head(5).index
top_routes_flights = delayed[delayed['Route'].isin(top_routes_list)]
weather_delays = top_routes_flights[top_routes_flights['WeatherDelay'] > 0]
print("Pейсы, погодные условия которых способствовали задержке",len(weather_delays))
```

## Вопрос 17. В каком месяце происходит наибольшее количество отмен по вине Carrier?

Май
Январь
Сентябрь
Апрель

```python
carrier_cancelled = flights_df[
    (flights_df['Cancelled'] == 1) &
    (flights_df['CancellationCode'] == 'A')
]
carrier_by_month = carrier_cancelled['Month'].value_counts().sort_index()
print(carrier_by_month)

months = {
    1: 'Январь',
    2: 'Февраль',
    3: 'Март',
    4: 'Апрель',
    5: 'Май',
    6: 'Июнь',
    7: 'Июль',
    8: 'Август',
    9: 'Сентябрь',
    10: 'Октябрь',
    11: 'Ноябрь',
    12: 'Декабрь'
}

print("Месяц, в котором происходит наибольшее количество отмен по вине Carrier", months[carrier_by_month.idxmax()])
```

## Вопрос 18. Определите перевозчика с наибольшим количеством отмен из-за перевозчика в соответствующем месяце из предыдущего вопроса.

9E
EV
HA
AA

```python
target_month = carrier_by_month.idxmax()
print(target_month)

target_data = flights_df[
    (flights_df['Cancelled'] == 1) &
    (flights_df['CancellationCode'] == 'A') &
    (flights_df['Month'] == target_month)
]

print("Перевозчика с наибольшим количеством отмен из-за перевозчика в соответствующем месяце", 
      target_data['UniqueCarrier'].value_counts().idxmax(), target_data['UniqueCarrier'].value_counts().max())
```
